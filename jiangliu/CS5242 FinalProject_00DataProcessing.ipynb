```python
import os
import random
import shutil
import re
import pandas as pd
## specify what type of files
from glob import glob 
pre = "/data/liu/Images"
split_pre = "/data/liu/dog_breeds"
# pre
```

```python
dirs = os.listdir(pre)
label_name = [name.split('-')[:2] for name in dirs]
```

```python
dog_breed_item = pd.DataFrame(label_name,columns=['breed_id','breed'])
dog_breed_item.head()
```

```python
dog_id = []
i=0
for breed in dirs:
    i+=1
    dog_id += glob(os.path.join(pre,breed,'*.jpg'))
```

```python
dog_breed_id = [dog.split('/')[-2:] for dog in dog_id]
```

```python
dog_breed_id = [dog[0].split('-')[:2]+[dog[1]] for dog in dog_breed_id]
dog_breed_item = pd.DataFrame(dog_breed_id,columns=['breed_id','breed','id'])
dog_breed_item.head()
```

```python
dog_breed_item.to_csv(os.path.join(split_pre,'all_dogs.csv'),index=False)
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(dog_breed_item, test_size=0.20,\
                                     random_state=5242, stratify=dog_breed_item['breed'])

print(len(df_train))
print(len(df_test))
```

```python
df_test.to_csv(os.path.join(split_pre,'df_test.csv'),index=False)
```

```python
df_train_train, df_val = train_test_split(df_train, test_size=0.20,\
                                     random_state=5242, stratify=df_train['breed'])
print(len(df_train_train))
print(len(df_val))
```

```python
df_train_train.to_csv(os.path.join(split_pre,'df_train.csv'),index=False)
```

```python
df_val.to_csv(os.path.join(split_pre,'df_val.csv'),index=False)
```

```python
dirs = os.listdir(pre)
# print(dirs)
for dirc in dirs:
    print('%s/%s' % (pre, dirc))
    dir = os.listdir('%s/%s/' % (pre, dirc))
    n = int(len(dir) * 0.2)
    sample = random.sample(dir, n)
#     print(sample)
    for name in sample:
        print('%s/%s/%s' % (pre, dirc, name))
        shutil.copy('%s/%s/%s' % (pre, dirc, name), '/data/liu/dog_breeds/test_images/')
        dir.remove(name)
    for name in dir:
        shutil.copy('%s/%s/%s' % (pre, dirc, name), '/data/liu/dog_breeds/train_images/')
```

```python
files = os.listdir('%s/train_images'%split_pre)
train_label = pd.DataFrame({'id':files})
train_label['breed_id'] = train_label['id'].apply(lambda x:str(x).split('_')[0])
train_label.to_csv('%s/%s'%(pre,'train_label.csv'), header=True,index=False)
train_label.head()
```

```python
files = os.listdir('%s/test_images'%split_pre)
test_label = pd.DataFrame({'id':files})
test_label['breed_id'] = test_label['id'].apply(lambda x:str(x).split('_')[0])
test_label.to_csv('%s/%s'%(pre,'test_label.csv'), header=True,index=False)
test_label.head()
```

```python
dirs = os.listdir(pre)
label_name = [name.split('-') for name in dirs]
label_name[:10]
```

```python
label_name_df = pd.DataFrame(label_name,columns=['breed_id','breed',1,2])[['breed_id','breed']]
label_name_df.head()
```

```python
train_label_name = pd.merge(train_label, label_name_df, how='left', on=['breed_id'])
# train_label.join(label_name_df,on='label',how='left')
train_label_name.to_csv('%s/%s'%(split_pre,'train_label_name.csv'), header=True,index=False)
train_label_name.head()
```

```python
test_label_name = pd.merge(test_label, label_name_df, how='left', on=['breed_id'])
# train_label.join(label_name_df,on='label',how='left')
test_label_name.to_csv('%s/%s'%(split_pre,'test_label_name.csv'), header=True,index=False)
test_label_name.head()
```
